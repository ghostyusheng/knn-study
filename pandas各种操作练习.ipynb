```python
def init():
    X = np.array([
        [1,2,3,4],
        [2,3,5,8],
        [8,2,1,3],
        [2,3,5,8],
        [8,2,1,3],
        [2,3,4,5]
    ])
    df = pd.DataFrame(X, index=dates, columns=list('ABCD'))
    return df
df = init()
df
```
